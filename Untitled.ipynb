```python
# To run this script, you need to install the following Python libraries.
# You can install them using pip:
# pip install opencv-python mediapipe pandas numpy scikit-learn matplotlib seaborn plotly requests xgboost shap scipy tensorflow networkx

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import xgboost as xgb
import shap
import warnings
import cv2
import mediapipe as mp
import time
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.sequence import pad_sequences
from scipy.spatial import Voronoi, voronoi_plot_2d
import networkx as nx

# Suppress specific warnings
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)

# --- 1. Data Generation and Helper Functions ---
def generate_synthetic_gaze_data_advanced(num_samples, class_type, random_seed=None):
    if random_seed is not None: np.random.seed(random_seed)
    if class_type == 'TD':
        avc, fas, asc = np.random.normal(0.15, 0.05, num_samples), np.random.normal(0.75, 0.10, num_samples), np.random.normal(0.70, 0.10, num_samples)
        fd_social, fd_nonsocial, pdr = np.random.normal(250, 30, num_samples), np.random.normal(100, 20, num_samples), np.random.normal(0.8, 0.15, num_samples)
        label = 0
    elif class_type == 'ASD':
        avc, fas, asc = np.random.normal(0.45, 0.10, num_samples), np.random.normal(0.25, 0.08, num_samples), np.random.normal(0.20, 0.08, num_samples)
        fd_social, fd_nonsocial, pdr = np.random.normal(120, 25, num_samples), np.random.normal(200, 30, num_samples), np.random.normal(0.3, 0.10, num_samples)
        label = 1
    else: raise ValueError("class_type must be 'TD' or 'ASD'")
    data = pd.DataFrame({'AOI_Vacancy_Counts': np.clip(avc, 0.01, 0.99), 'Favorable_AOI_Shifts': np.clip(fas, 0.01, 0.99), 'AOI_Switch_Counts': np.clip(asc, 0.01, 0.99), 'Fixation_Duration_Social': np.clip(fd_social, 50, 400), 'Fixation_Duration_NonSocial': np.clip(fd_nonsocial, 50, 400), 'Pupil_Dilation_Response': np.clip(pdr, 0.01, 1.0), 'Class': label})
    return data

def calculate_social_gaze_ratio(df):
    sum_durations = df['Fixation_Duration_Social'] + df['Fixation_Duration_NonSocial']
    df['Social_Gaze_Ratio'] = df['Fixation_Duration_Social'] / (sum_durations + 1e-9)
    df['Social_Gaze_Ratio'] = np.nan_to_num(df['Social_Gaze_Ratio'], nan=0.0, posinf=0.0, neginf=0.0)
    return df

def detect_fixations(gaze_points, max_dist=35, min_duration_ms=100):
    if not gaze_points: return []
    fixations = []
    current_fixation_points = [gaze_points[0]]
    for i in range(1, len(gaze_points)):
        dist = np.linalg.norm(np.array(gaze_points[i][1:]) - np.array(current_fixation_points[-1][1:]))
        if dist <= max_dist:
            current_fixation_points.append(gaze_points[i])
        else:
            duration_ms = (current_fixation_points[-1][0] - current_fixation_points[0][0]) * 1000
            if duration_ms >= min_duration_ms:
                x_center = int(np.mean([p[1] for p in current_fixation_points]))
                y_center = int(np.mean([p[2] for p in current_fixation_points]))
                fixations.append((x_center, y_center, duration_ms))
            current_fixation_points = [gaze_points[i]]
    return fixations

def plot_scanpath(image, fixations):
    scanpath_img = image.copy()
    if not fixations: return scanpath_img
    for i in range(len(fixations) - 1):
        cv2.line(scanpath_img, (fixations[i][0], fixations[i][1]), (fixations[i+1][0], fixations[i+1][1]), (0, 0, 255), 2)
    for i, (x, y, duration) in enumerate(fixations):
        radius = int(5 + (duration / 1000) * 50)
        cv2.circle(scanpath_img, (x, y), radius, (0, 255, 0), 2)
        cv2.putText(scanpath_img, str(i + 1), (x + 5, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
    return scanpath_img

# --- [NEW] Comprehensive Visualization Function ---
def generate_master_visualization_suite(dataset):
    print("\n--- Generating Master Suite of Data Visualizations ---")
    plt.style.use('seaborn-v0_8-whitegrid')
    custom_palette = {'Typically Developing': '#06D6A0', 'Autism Spectrum Disorder': '#FF6B6B'}
    features = ['AOI_Vacancy_Counts', 'Favorable_AOI_Shifts', 'Social_Gaze_Ratio', 'Pupil_Dilation_Response']
    
    # 1. Basic Distribution Plots
    print("Generating: 1. Basic Distribution Plots")
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    sns.histplot(data=dataset, x='Social_Gaze_Ratio', hue='Class_Label', kde=True, palette=custom_palette, ax=axes[0, 0]).set_title('Histogram & Density')
    sns.boxplot(data=dataset, x='Class_Label', y='Social_Gaze_Ratio', palette=custom_palette, ax=axes[0, 1]).set_title('Box Plot')
    sns.violinplot(data=dataset, x='Class_Label', y='Social_Gaze_Ratio', palette=custom_palette, ax=axes[1, 0]).set_title('Violin Plot')
    sns.stripplot(data=dataset, x='Class_Label', y='Social_Gaze_Ratio', palette=custom_palette, ax=axes[1, 1]).set_title('Strip Plot')
    plt.tight_layout(); plt.show()

    # 2. Relationship & Regression Plots
    print("Generating: 2. Relationship & Regression Plots")
    sns.lmplot(data=dataset, x='AOI_Vacancy_Counts', y='Favorable_AOI_Shifts', hue='Class_Label', palette=custom_palette, line_kws={'lw': 2}, scatter_kws={'alpha': 0.4})
    plt.title('Linear Regression Plot'); plt.show()
    fig = px.scatter(dataset, x='AOI_Vacancy_Counts', y='Social_Gaze_Ratio', color='Class_Label', size='Pupil_Dilation_Response', color_discrete_map=custom_palette, title='Bubble Plot (3 Variables)')
    fig.show()

    # 3. Composition Plots
    print("Generating: 3. Composition Plots")
    class_counts = dataset['Class_Label'].value_counts()
    fig = go.Figure(data=[go.Pie(labels=class_counts.index, values=class_counts.values, hole=.4, marker_colors=['#06D6A0', '#FF6B6B'])])
    fig.update_layout(title_text='Donut Chart of Class Distribution'); fig.show()

    # 4. Correlation & Multivariate Plots
    print("Generating: 4. Correlation & Multivariate Plots")
    correlation_matrix = dataset[features].corr()
    plt.figure(figsize=(8, 6)); sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f").set_title('Correlogram (Correlation Heatmap)'); plt.show()
    pc_features = ['AOI_Vacancy_Counts', 'Favorable_AOI_Shifts', 'Social_Gaze_Ratio', 'Pupil_Dilation_Response', 'Class']
    fig = px.parallel_coordinates(dataset, dimensions=pc_features, color='Class', color_continuous_scale=px.colors.sequential.Plasma, title='Parallel Coordinates Plot')
    fig.show()
    
    # 5. Hierarchical & Flow Plots (Conceptual)
    print("Generating: 5. Hierarchical & Flow Plots")
    tree_data = pd.DataFrame({'Category': ['Social Gaze', 'Social Gaze', 'Gaze Dynamics', 'Attention Focus'], 'Metric': ['Fixation_Duration_Social', 'Social_Gaze_Ratio', 'Favorable_AOI_Shifts', 'AOI_Vacancy_Counts'], 'Contribution': [0.4, 0.3, 0.2, 0.1]})
    fig = px.treemap(tree_data, path=['Category', 'Metric'], values='Contribution', title='Tree Chart (Treemap) of Feature Categories')
    fig.show()
    fig = go.Figure(data=[go.Sankey(node=dict(pad=15, thickness=20, label=["Social AOI", "Non-Social AOI", "Vacancy", "TD Group", "ASD Group"]),link=dict(source=[0, 0, 1, 1, 2, 2], target=[3, 4, 3, 4, 3, 4], value=[40, 15, 10, 30, 5, 10]))])
    fig.update_layout(title_text="Sankey Diagram: Conceptual Gaze Flow"); fig.show()

    # 6. Advanced Spatial & 3D Plots
    print("Generating: 6. Advanced Spatial & 3D Plots")
    fig_3d = px.scatter_3d(dataset.sample(min(500, len(dataset))), x='AOI_Vacancy_Counts', y='Favorable_AOI_Shifts', z='Social_Gaze_Ratio', color='Class_Label', color_discrete_map=custom_palette, title='Interactive 3D Scatter Plot')
    fig_3d.show()
    
    scaler = StandardScaler(); pca = PCA(n_components=2)
    principal_components = pca.fit_transform(scaler.fit_transform(dataset[features]))
    vor = Voronoi(principal_components)
    fig, ax = plt.subplots(figsize=(10, 8))
    voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='orange')
    sns.scatterplot(x=principal_components[:, 0], y=principal_components[:, 1], hue=dataset['Class_Label'], palette=custom_palette, ax=ax)
    plt.title('Voronoi Diagram of PCA-Reduced Data'); plt.show()

# --- Main Application Class ---
class EyeTrackerApp:
    def __init__(self, stimulus_path, data_path='autistic_eye_movement_data.csv'):
        self.stimulus_path = stimulus_path
        self.data_path = data_path
        self.stimulus_img = cv2.imread(stimulus_path)
        if self.stimulus_img is None: raise FileNotFoundError(f"Stimulus image not found at {stimulus_path}")
        self.cam = cv2.VideoCapture(0)
        self.face_mesh = mp.solutions.face_mesh.FaceMesh(refine_landmarks=True, max_num_faces=1)
        self.aois, self.gaze_log, self.xgb_model, self.lstm_model, self.scaler = {}, [], None, None, None
        self.features = ['AOI_Vacancy_Counts', 'Favorable_AOI_Shifts', 'AOI_Switch_Counts', 'Fixation_Duration_Social', 'Fixation_Duration_NonSocial', 'Pupil_Dilation_Response', 'Social_Gaze_Ratio']
        self.sequence_length = 100

    def _mouse_callback(self, event, x, y, flags, param):
        if event == cv2.EVENT_LBUTTONDOWN:
            aoi_name, point_name = param['current_step']
            self.aois[aoi_name][point_name] = (x, y)
            param['click_count'] += 1

    def calibrate_aois(self):
        print("--- Interactive AOI Calibration ---")
        self.aois = {'social': {'tl': None, 'br': None}, 'nonsocial': {'tl': None, 'br': None}}
        steps = [('social', 'tl'), ('social', 'br'), ('nonsocial', 'tl'), ('nonsocial', 'br')]
        instructions = {'social_tl': "Click the TOP-LEFT corner of the FACE", 'social_br': "Click the BOTTOM-RIGHT corner of the FACE", 'nonsocial_tl': "Click the TOP-LEFT corner of the OBJECT", 'nonsocial_br': "Click the BOTTOM-RIGHT corner of the OBJECT"}
        param = {'click_count': 0}
        cv2.namedWindow("AOI Calibration")
        while param['click_count'] < 4:
            param['current_step'] = steps[param['click_count']]
            cv2.setMouseCallback("AOI Calibration", self._mouse_callback, param)
            display_img = self.stimulus_img.copy()
            cv2.putText(display_img, instructions[f"{param['current_step'][0]}_{param['current_step'][1]}"], (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            cv2.imshow("AOI Calibration", display_img)
            if cv2.waitKey(1) & 0xFF == ord('q'): break
        cv2.destroyWindow("AOI Calibration")
        print("AOI Calibration Complete.")

    def run_test(self, duration=20):
        print(f"\n--- Running Gaze Test for {duration} seconds ---")
        self.gaze_log = []
        start_time = time.time()
        while time.time() - start_time < duration:
            ret, frame = self.cam.read()
            if not ret: break
            frame = cv2.flip(frame, 1)
            results = self.face_mesh.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            stim_display = self.stimulus_img.copy()
            cv2.rectangle(stim_display, self.aois['social']['tl'], self.aois['social']['br'], (0, 255, 0), 2)
            cv2.rectangle(stim_display, self.aois['nonsocial']['tl'], self.aois['nonsocial']['br'], (0, 0, 255), 2)
            if results.multi_face_landmarks:
                landmarks = results.multi_face_landmarks[0].landmark
                gaze_x, gaze_y = 1 - landmarks[468].x, landmarks[468].y
                screen_x, screen_y = int(gaze_x * self.stimulus_img.shape[1]), int(gaze_y * self.stimulus_img.shape[0])
                self.gaze_log.append((time.time() - start_time, screen_x, screen_y))
                cv2.circle(stim_display, (screen_x, screen_y), 10, (255, 0, 0), -1)
            cv2.imshow("Stimulus", stim_display)
            if cv2.waitKey(1) & 0xFF == ord('q'): break
        cv2.destroyAllWindows()
        print("Test complete.")

    def _create_lstm_model(self, input_shape):
        model = Sequential([LSTM(64, input_shape=input_shape, return_sequences=True), Dropout(0.3), LSTM(32), Dropout(0.3), Dense(16, activation='relu'), Dense(1, activation='sigmoid')])
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        return model

    def train_models(self, random_seed=42):
        print("\n--- Training Models ---")
        try:
            asd_df = pd.read_csv(self.data_path)
            td_df = generate_synthetic_gaze_data_advanced(len(asd_df), 'TD', random_seed)
            self.full_dataset = pd.concat([asd_df.assign(Class=1, Class_Label='Autism Spectrum Disorder'), td_df.assign(Class=0, Class_Label='Typically Developing')], ignore_index=True)
        except FileNotFoundError:
            asd_df = generate_synthetic_gaze_data_advanced(700, 'ASD', random_seed + 1)
            td_df = generate_synthetic_gaze_data_advanced(700, 'TD', random_seed)
            self.full_dataset = pd.concat([asd_df.assign(Class=1, Class_Label='Autism Spectrum Disorder'), td_df.assign(Class=0, Class_Label='Typically Developing')], ignore_index=True)
        
        self.full_dataset = calculate_social_gaze_ratio(self.full_dataset)
        X, y = self.full_dataset[self.features], self.full_dataset['Class']
        
        self.scaler = StandardScaler().fit(X)
        self.xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=random_seed).fit(self.scaler.transform(X), y)
        print("XGBoost model trained.")
        
        sequences = [np.random.rand(np.random.randint(50, 150), 2) for _ in range(len(self.full_dataset))]
        padded_sequences = pad_sequences(sequences, maxlen=self.sequence_length, dtype='float32', padding='post', truncating='post')
        self.lstm_model = self._create_lstm_model((self.sequence_length, 2))
        self.lstm_model.fit(padded_sequences, y, epochs=10, batch_size=32, verbose=0)
        print("LSTM model trained.")
        return self.full_dataset

    def analyze_and_predict(self):
        print("\n--- Analyzing Your Gaze Data ---")
        summary_features = self._calculate_summary_features()
        xgb_prediction_proba = self.xgb_model.predict_proba(self.scaler.transform(summary_features))[0]
        
        gaze_sequence = np.array([[x, y] for _, x, y in self.gaze_log])
        gaze_sequence_normalized = (gaze_sequence - np.mean(gaze_sequence, axis=0)) / (np.std(gaze_sequence, axis=0) + 1e-9)
        padded_sequence = pad_sequences([gaze_sequence_normalized], maxlen=self.sequence_length, dtype='float32', padding='post', truncating='post')
        lstm_prediction_proba = self.lstm_model.predict(padded_sequence, verbose=0)[0]
        
        # Display Scanpath
        fixations = detect_fixations(self.gaze_log)
        scanpath_img = plot_scanpath(self.stimulus_img.copy(), fixations)
        plt.figure(figsize=(12, 8)); plt.imshow(cv2.cvtColor(scanpath_img, cv2.COLOR_BGR2RGB)); plt.title("Your Gaze Scanpath"); plt.axis('off'); plt.show()
        
        print("\n" + "="*50 + "\nFINAL CONCLUSION\n" + "="*50)
        print("--- XGBoost Model (on Summary Features) ---")
        print(f"Predicted Class: {'Autism Spectrum Disorder' if xgb_prediction_proba[1] > 0.5 else 'Typically Developing'}")
        print(f"Model Confidence (TD, ASD): [{xgb_prediction_proba[0]:.2f}, {xgb_prediction_proba[1]:.2f}]")
        print("\n--- LSTM Model (on Gaze Sequence) ---")
        print(f"Predicted Class: {'Autism Spectrum Disorder' if lstm_prediction_proba[0] > 0.5 else 'Typically Developing'}")
        print(f"Model Confidence (TD, ASD): [{1-lstm_prediction_proba[0]:.2f}, {lstm_prediction_proba[0]:.2f}]")
        
        print("\n--- SHAP Explanation for XGBoost Model ---")
        explainer = shap.TreeExplainer(self.xgb_model)
        shap_values = explainer.shap_values(self.scaler.transform(summary_features))
        shap.summary_plot(shap_values, features=summary_features, plot_type="bar", show=False)
        plt.title("SHAP Feature Importance for Your Prediction"); plt.show()

    def _calculate_summary_features(self):
        time_in = {'social': 0, 'nonsocial': 0, 'vacancy': 0}
        last_time = 0
        for t, x, y in self.gaze_log:
            dt = t - last_time; last_time = t
            if self.aois['social']['tl'][0] < x < self.aois['social']['br'][0] and self.aois['social']['tl'][1] < y < self.aois['social']['br'][1]: time_in['social'] += dt
            elif self.aois['nonsocial']['tl'][0] < x < self.aois['nonsocial']['br'][0] and self.aois['nonsocial']['tl'][1] < y < self.aois['nonsocial']['br'][1]: time_in['nonsocial'] += dt
            else: time_in['vacancy'] += dt
        total_time = sum(time_in.values()) or 1
        summary = pd.DataFrame([{"Fixation_Duration_Social": time_in['social'] * 1000, "Fixation_Duration_NonSocial": time_in['nonsocial'] * 1000, "AOI_Vacancy_Counts": time_in['vacancy'] / total_time, "Pupil_Dilation_Response": 0.5, "Favorable_AOI_Shifts": 0.5, "AOI_Switch_Counts": 0.5}])
        return calculate_social_gaze_ratio(summary)[self.features]

if __name__ == "__main__":
    try:
        app = EyeTrackerApp(stimulus_path="stimulus.jpg")
        app.calibrate_aois()
        training_data = app.train_models()
        
        generate_master_visualization_suite(training_data)
        
        app.run_test(duration=20)
        app.analyze_and_predict()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        print("Please ensure your webcam is connected and all files ('stimulus.jpg', 'autistic_eye_movement_data.csv') are in the correct folder.")
    finally:
        if 'app' in locals() and app.cam.isOpened():
            app.cam.release()
        cv2.destroyAllWindows()
        print("\nApplication finished.")
```
